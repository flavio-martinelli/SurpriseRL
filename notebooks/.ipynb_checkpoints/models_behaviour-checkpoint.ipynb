# Show behaviour of neuron models

Show how to define the models and plot their behaviour.
The different models are in neuron_models.py:
- **Lif**: Leaky Integrate and Fire neuron *'layer'* trainable through gradient descent, with deterministic threshold
- **Lif3f**: Leaky Integrate and Fire EI mismatch *'network'* implementing 3 factor learning rules, with stochastic threshold
- **Lif3fLoop**: Like the one above, but not wrapped in an AbstractRNNCell

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow.keras as keras

# My imports
from neuron_models import Lif, Lif3f, Lif3fLoop
from plot_utils import raster_plot, v_plot

# set generic params
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # removes annoying warnings
tf.config.experimental_run_functions_eagerly(True) # allows debugging on @tf.function decorated funs

time_steps = 100
```

### Lif

#### Single neuron, single input

```python
n_inputs = 1

inputs = np.zeros([1, time_steps, n_inputs])
inputs[0, (10, 30, 80), 0] = 1.0
inputs = tf.Variable(inputs, dtype=tf.float32)

cell1 = Lif(n_rec=1, tau=10, thr=1, dt=1,
             w_in=[[5.0]],
             w_rec=[[0.0]],
             track_v=True)

rnn1 = keras.layers.RNN(cell1, return_sequences=True, return_state=False)
spk_out, v_out = tf.unstack(rnn1(inputs), num=2, axis=-1)


f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 4), gridspec_kw={'height_ratios': [1, 6]})

raster_plot(ax1, inputs[0], linewidth = 1, color='r')
ax1.set_title('Input spike train')
ax1.axes.get_yaxis().set_visible(False)
ax1.axes.get_xaxis().set_visible(False)

v_plot(ax2, v_out[0], out_spikes=spk_out[0], linewidth=1, color='b')
ax2.set_title(f"Membrane potential")
ax2.set_xlabel('time step')
ax2.set_ylabel('voltage [unitless]')
ax2.legend([f'tau={cell1.tau.numpy()}, decay={cell1._decay:.3f}'])
ax2.locator_params(axis='y', nbins=10)
ax2.grid(which='minor')
ax2.grid()
```

#### Single neuron, multiple inputs (of different weights)

```python
n_inputs = 3

inputs = np.zeros([1, time_steps, n_inputs])
inputs[0, (10, 15, 20, 30), 0] = 1.0
inputs[0, (40, 45, 50, 60, 62), 1] = 1.0
inputs[0, (70, 90), 2] = 1.0

inputs = tf.Variable(inputs, dtype=tf.float32)

cell1 = Lif(n_rec=1, tau=10, thr=1, dt=1,
             w_in=[[7.0], [-3.0], [2.0]],
             w_rec=[[0.0]],
             track_v=True)

rnn1 = keras.layers.RNN(cell1, return_sequences=True, return_state=False)
spk_out, v_out = tf.unstack(rnn1(inputs), num=2, axis=-1)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 4), gridspec_kw={'height_ratios': [1, 4]})

raster_plot(ax1, inputs[0], linewidth = 1, color='r')
ax1.set_title('Input spike train')
ax1.axes.get_yaxis().set_visible(False)
ax1.axes.get_xaxis().set_visible(False)

v_plot(ax2, v_out[0], out_spikes=spk_out[0], linewidth=1, color='b')
ax2.set_title(f"Membrane potential")
ax2.set_xlabel('time step')
ax2.set_ylabel('voltage [unitless]')
ax2.legend([f'tau={cell1.tau.numpy()}, decay={cell1._decay:.3f}'])
ax2.locator_params(axis='y', nbins=10)
ax2.grid(which='minor')
ax2.grid()
```

#### Multiple neuron with different taus and multiple inputs

```python
n_inputs = 3

inputs = np.zeros([1, time_steps, n_inputs])
inputs[0, (10, 30), 0] = 1.0
inputs[0, (40, 60), 1] = 1.0
inputs[0, (70, 90), 2] = 1.0

inputs = tf.Variable(inputs, dtype=tf.float32)

cell1 = Lif(n_rec=1, tau=10, thr=1, dt=1,
             w_in=[[5.0], [7.5], [7.5]],
             w_rec=[[0.0]],
             track_v=True)

cell2 = Lif(n_rec=1, tau=15, thr=1, dt=1,
             w_in=[[5.0], [7.5], [7.5]],
             w_rec=[[0.0]],
             track_v=True)

cell3 = Lif(n_rec=1, tau=20, thr=1, dt=1,
             w_in=[[5.0], [7.5], [7.5]],
             w_rec=[[0.0]],
             track_v=True)

rnn1 = keras.layers.RNN(cell1, return_sequences=True, return_state=False)
spk_out, v_out = tf.unstack(rnn1(inputs), num=2, axis=-1)

rnn2 = keras.layers.RNN(cell2, return_sequences=True, return_state=False)
spk_out2, v_out2 = tf.unstack(rnn2(inputs), num=2, axis=-1)

rnn3 = keras.layers.RNN(cell3, return_sequences=True, return_state=False)
spk_out3, v_out3 = tf.unstack(rnn3(inputs), num=2, axis=-1)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 4), gridspec_kw={'height_ratios': [1, 4]})

raster_plot(ax1, inputs[0], linewidth = 1, color='r')
ax1.set_title('Input spike train')
ax1.axes.get_yaxis().set_visible(False)
ax1.axes.get_xaxis().set_visible(False)

v_plot(ax2, v_out[0], out_spikes=spk_out[0], linewidth=1, color='b')
v_plot(ax2, v_out2[0], out_spikes=spk_out2[0], linewidth=1, color='g')
v_plot(ax2, v_out3[0], out_spikes=spk_out3[0], linewidth=1, color='m')
ax2.set_title(f"Membrane potential")
ax2.set_xlabel('time step')
ax2.set_ylabel('voltage [unitless]')
ax2.legend([f'tau={cell1.tau.numpy()}, decay={cell1._decay:.3f}',
            f'tau={cell2.tau.numpy()}, decay={cell2._decay:.3f}',
            f'tau={cell3.tau.numpy()}, decay={cell3._decay:.3f}'])
ax2.locator_params(axis='y', nbins=10)
ax2.grid(which='minor')
ax2.grid()
```

### Lif3f

Change the seed to see the stochastic thresholds change

```python
# tf.config.experimental_run_functions_eagerly(False) # allows debugging on @tf.function decorated funs

n_inputs = 2  # Minimum required because the network needs 2 differnt input populations 
seed = 1

inputs = np.zeros([1, time_steps, n_inputs])
inputs[0, (10), 0] = 1.0
inputs[0, (70), 1] = 1.0
inputs = tf.Variable(inputs, dtype=tf.float32)

cell1 = Lif3f(n_rec=2, tau_v=10, dt=1,
              w_in=[[0.5, -0.3], [1000, 0.5]],
              track_vars=True, seed=seed)

rnn = keras.layers.RNN(cell1, return_sequences=True, return_state=False)

out = rnn(inputs)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 4), gridspec_kw={'height_ratios': [1, 4]})

raster_plot(ax1, inputs[0], linewidth = 1, color='r')
ax1.set_title('Input spike train')
ax1.axes.get_yaxis().set_visible(False)
ax1.axes.get_xaxis().set_visible(False)

v_plot(ax2, out['state'].v[0], out_spikes=out['state'].z[0], linewidth=1, color='b')
ax2.set_title(f"Membrane potential")
ax2.set_xlabel('time step')
ax2.axes.get_yaxis().set_visible(False)
ax2.locator_params(axis='y', nbins=10)
ax2.grid()
```

In this network the weights from the second input (working memory) are clipped to be negative if projecting to the first error population and positive if projecting to the second error population, hence the clipping of such extreme weight values we see above

### Lif3fLoop


```python
n_inputs = 2  # Minimum required because the network needs 2 differnt input populations 
seed = 128

inputs = np.zeros([1, time_steps, n_inputs])
inputs[0, (10), 0] = 1.0
inputs[0, (70), 1] = 1.0
inputs = tf.Variable(inputs, dtype=tf.float32)

last_v = None
last_activity = None

snn = Lif3fLoop(n_in=2, n_rec=2, tau_v=10, dt=1,
               w_in=[[0.40, -0.3], [-0.3, 0.46]], seed=seed)

@tf.function
def run(inputs, last_v, last_activity):
    return snn.run(inputs[0], last_v, last_activity)

v, z, activity, surprise = run(inputs, last_v, last_activity)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 4), gridspec_kw={'height_ratios': [1, 4]})

raster_plot(ax1, inputs[0], linewidth = 1, color='r')
ax1.set_title('Input spike train')
ax1.axes.get_yaxis().set_visible(False)
ax1.axes.get_xaxis().set_visible(False)

v_plot(ax2, v, out_spikes=z, linewidth=1, color='b')
ax2.set_title(f"Membrane potential")
ax2.set_xlabel('time step')
ax2.axes.get_yaxis().set_visible(False)
ax2.locator_params(axis='y', nbins=10)
ax2.grid()
```
