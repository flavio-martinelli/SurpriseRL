# Maze construction

Initialize Maze2D with these parameters

- size: tuple of 2 elements for the size of the maze
- walls: False allows the agent to pop up from the other size of the maze instead of hitting a wall
- moves_list: custom list of moves (as list of 2d lists), the default moves_list is [[0,1], [0,-1], [1,0], [-1,0]] # right, left, down, up

```python
m = Maze2D(size=(3,6), walls=False)
# pprint(vars(m))
```

# Transition matrix with walls and without

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
sizes = [(2,2), (4,4), (6,6)]

for i,size in enumerate(sizes):
    maze = Maze2D(size=size, walls=False)
    pcol = maze.plot_t_mat(ax[i])
    pcol.set_clim(0.0,0.5)
    ax[i].set_title(f"Maze of dimensions: {maze.size} " + f"with{'out' if maze.walls is False else ''} walls")

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
cbar = fig.colorbar(pcol, cax=cbar_ax)
plt.show()
```

## Transition matrix with explicit actions

```python
maze = Maze2D(size=(4,4), walls=False)
t_mat = maze.build_transition_matrix(separate_actions=True)
t_mat.shape
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 4))

for i,mat in enumerate(t_mat):
    pcol = maze.plot_t_mat(ax[i], t_mat=mat)
    pcol.set_clim(0.0,0.5)
    ax[i].set_title(f"Maze of dimensions {maze.size}, " + f"move #{i}")
    
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
cbar = fig.colorbar(pcol, cax=cbar_ax)
plt.show()
```

##### Comparing the transition matrix without actions and the sum of all above transition matrices

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

pcol = maze.plot_t_mat(ax[0])
pcol.set_clim(0.0,0.5)
ax[0].set_title(f"full matrix")

pcol = maze.plot_t_mat(ax[1], t_mat=np.sum(t_mat, axis=0))
pcol.set_clim(0.0,0.5)
ax[1].set_title(f"sum of different actions")

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
cbar = fig.colorbar(pcol, cax=cbar_ax)
plt.show()
```

## Actions

```python
size = (4,4)

maze = Maze2D(size, walls=False, proba_opposite=0.30, proba_others=0.05)
maze._state = maze.maze_array[1,2]

fig, ax = mplot.base_fig(size)
mplot.draw_cell_numbers(ax, maze.maze_array)
mplot.color_square(ax, maze.get_coordinates(maze._state), 'b')
```

#### Actions with stochasticity, try run the cell multiple times

```python
print(maze.moves_list)
```

```python
action = 2 # go down action ([1,0])

_ , next_supposed_position = maze.move_in_maze(maze._state, maze.moves_list[action])

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ax in axes:
    mplot.base_fig(size, ax)
    mplot.draw_cell_numbers(ax, maze.maze_array)
    transition = maze._step(2)
    new_position = maze.get_coordinates(transition.observation)
    mplot.color_square(ax, new_position, 'b')
    mplot.draw_arrow(ax, maze.get_coordinates(maze._state), next_supposed_position, color='r')
```

```python
rec=[]
maze.move_reward(0)
for i in range(1000):
    maze._state = 6
    transition = maze._step(2)
    rec.append(transition.observation)
    maze._state = 6

plt.hist(np.array(rec))
plt.title("next cell position after choosing action 'down' from cell 6")
plt.show()
```

#### Agent always choosing to go right: animation


```python
from IPython.display import Video, HTML

size = (4,10)

maze = Maze2D(size, walls=False, proba_opposite=0.15, proba_others=0.05)
maze._state = maze.maze_array[1,2]

action = 0 # go right action ([0,1])

fig, ax = mplot.base_fig(size)
mplot.draw_cell_numbers(ax, maze.maze_array)
mplot.color_square(ax, maze.get_coordinates(maze._state), 'b')
mplot.color_square(ax, maze.get_coordinates(maze.reward_pos), 'r')
camera = Camera(fig)

for i in range(20):
    _ , next_supposed_position = maze.move_in_maze(maze._state, maze.moves_list[action])
    
    plt.draw()
    camera.snap()
    
    mplot.base_fig(size, ax)
    mplot.draw_cell_numbers(ax, maze.maze_array)
    mplot.color_square(ax, maze.get_coordinates(maze._state), 'b')
    mplot.color_square(ax, maze.get_coordinates(maze.reward_pos), 'r')
    mplot.draw_arrow(ax, maze.get_coordinates(maze._state), next_supposed_position, color='r')
    
    transition = maze._step(action)
    next_position = transition.observation

    plt.draw()
    camera.snap()
    
    mplot.base_fig(size, ax)
    mplot.draw_cell_numbers(ax, maze.maze_array)
    mplot.color_square(ax, maze.get_coordinates(maze._state), 'b')
    mplot.color_square(ax, maze.get_coordinates(maze.reward_pos), 'r')
    
    if i==10:
        action = 2

plt.draw()
camera.snap()
plt.close()

animation = camera.animate(interval=500)
HTML(animation.to_html5_video())
```

### Transition matrix with probabilistic actions

```python
maze = Maze2D(size=(4,4), walls=False, proba_opposite=0.15, proba_others=0.05)
t_mat = maze.build_transition_matrix(separate_actions=True)

fig, ax = plt.subplots(1, 4, figsize=(20, 4))

for i,mat in enumerate(t_mat):
    pcol = maze.plot_t_mat(ax[i], t_mat=mat)
    pcol.set_clim(0.0,0.5)
    ax[i].set_title(f"Maze of dimensions {maze.size}, " + f"move #{i}")
    
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
cbar = fig.colorbar(pcol, cax=cbar_ax)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

pcol = maze.plot_t_mat(ax[0], maze.build_transition_matrix())
pcol.set_clim(0.0,0.5)
ax[0].set_title(f"full matrix")

pcol = maze.plot_t_mat(ax[1], t_mat=np.sum(t_mat, axis=0))
pcol.set_clim(0.0,0.5)
ax[1].set_title(f"sum of different actions")

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
cbar = fig.colorbar(pcol, cax=cbar_ax)
plt.show()
```

## Reward 

```python
size = (6,7)

maze = Maze2D(size, walls=False)
maze.agent_pos = maze.maze_array[1,1]
maze.move_reward(maze.maze_array[1,2])

fig, ax = mplot.base_fig(size)
mplot.draw_cell_numbers(ax, maze.reward_array)
mplot.color_square(ax, maze.get_coordinates(maze.agent_pos), 'b')
mplot.color_square(ax, maze.get_coordinates(maze.reward_pos), 'r')
```

#### Set up wall penalties

```python
reward_env = np.zeros_like(maze.maze_array, dtype=float)
reward_env[(0,-1),:] = -0.1
reward_env[:,(0,-1)] = -0.1
reward_env
```

```python
maze.env_reward = reward_env
maze.move_reward(maze.maze_array[4,3])
maze._state = maze.maze_array[4,2]


fig, (ax1,ax2) = plt.subplots(1,2,figsize=(14,6))
mplot.base_fig(size, ax1)
mplot.draw_cell_numbers(ax1, maze.reward_array)
mplot.color_square(ax1, maze.get_coordinates(maze.agent_pos), 'b')
mplot.color_square(ax1, maze.get_coordinates(maze.reward_pos), 'r')
ax1.set_title('Maze showing reward values')

mplot.base_fig(size, ax2)
mplot.draw_cell_numbers(ax2, maze.maze_array)
mplot.color_square(ax2, maze.get_coordinates(maze.agent_pos), 'b')
mplot.color_square(ax2, maze.get_coordinates(maze.reward_pos), 'r')
ax2.set_title('Maze showing cell numbers')
```

```python
for action in range(4):
    maze._state=30
    transition = maze._step(action)
    next_cell, reward = transition.observation[0], transition.reward
    print(f"agent moved from cell 30 to cell {next_cell} and received reward={reward}")
```
