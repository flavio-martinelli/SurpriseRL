```python
from bayes_opt import BayesianOptimization
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import mlab
from matplotlib import gridspec
%matplotlib inline


def target(x, y):
    a = np.exp(-( (x - 2)**2/0.7 + (y - 4)**2/1.2) + (x - 2)*(y - 4)/1.6 )
    b = np.exp(-( (x - 4)**2/3 + (y - 2)**2/2.) )
    c = np.exp(-( (x - 4)**2/0.5 + (y - 4)**2/0.5) + (x - 4)*(y - 4)/0.5 )
    d = np.sin(3.1415 * x)
    e = np.exp(-( (x - 5.5)**2/0.5 + (y - 5.5)**2/.5) )
    return 2*a + b - c + 0.17 * d + 2*e


n = 1e5
x = y = np.linspace(0, 6, 300)
X, Y = np.meshgrid(x, y)
x = X.ravel()
y = Y.ravel()
X = np.vstack([x, y]).T[:, [1, 0]]
z = target(x, y)


fig, axis = plt.subplots(1, 1, figsize=(14, 10))
gridsize=150

im = axis.hexbin(y, x, C=z, gridsize=gridsize, cmap=cm.jet, bins=None, vmin=-0.9, vmax=2.1)
axis.axis([x.min(), x.max(), y.min(), y.max()])

cb = fig.colorbar(im, )
cb.set_label('Value')


def posterior(bo, X):
    bo._gp.fit(bo._space.params, bo._space.target)
    mu, sigma = bo._gp.predict(X, return_std=True)
    return mu, sigma

def plot_2d(name=None):

    mu, s = posterior(bo, X)

    print(mu, s)
    
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))
    gridsize=150

    # GP regression output
    ax[0].set_title('Gaussian Process Predicted Mean', fontdict={'size':15})
    im00 = ax[0].hexbin(x, y, C=mu, gridsize=gridsize, cmap=cm.jet, bins=None)
    ax[0].axis([x.min(), x.max(), y.min(), y.max()])
    ax[0].plot(bo._space.params[:, 1], bo._space.params[:, 0], 'D', markersize=4, color='k', label='Observations')

    ax[1].set_title('Gausian Process Variance', fontdict={'size':15})
    im01 = ax[1].hexbin(x, y, C=s, gridsize=gridsize, cmap=cm.jet, bins=None)
    ax[1].axis([x.min(), x.max(), y.min(), y.max()])

    for im, axis in zip([im00, im01], ax.flatten()):
        cb = fig.colorbar(im, ax=axis)
        # cb.set_label('Value')

    if name is None:
        name = '_'

    plt.tight_layout()

    # Save or show figure?
    # fig.savefig('bo_eg_' + name + '.png')
    plt.show()

bo = BayesianOptimization(target, {'x': (0, 6), 'y': (0, 6)})

bo.maximize(init_points=3, n_iter=0, acq='ucb', kappa=10)

plot_2d()
```
